# Übung: Glätten


Aufgabe:

* Glätten eines verrauschten Signals
* Reduzieren der Datenpunkte

Tipps:

* Überprüfen des konstanten Abstands der Punkte (`df.resample("d").mean()`)
* Reduzieren der Datenpunkte durch Resamplen.
* Größe des Plots anpassen, Kurzform:

~~~
df.plot(figsize=(10,5))
~~~

* Größe des Plots anpassen, Langform:

~~~
#dpi controls relative size of font
fig = plt.figure(figsize=(6,3), dpi=150)
# create an ax object 
ax = plt.axes() 
# don't forget to tell pandas.plot() about new ax object
dg[['mean10', 'mean30']].plot(ax=ax) 
~~~

```python
%reset -f
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

fn = 'data/demo_tdaily_imp.csv'
df = pd.read_csv(fn) 
# short form
df = pd.read_csv(fn, parse_dates=[0], index_col=0)
df = df[['mydat']]
print(df.shape)
df.head(3)
```

```python
new_df = df.resample('D').mean()
display(new_df)
```

```python
import plotly_express as px
new_df = df.resample('D').mean()
display(new_df)
plot = px.line(new_df)
plot.show()
```

```python
new_df = df.resample('D').mean()
display(new_df)
window_size = 10
new_df[f'window_{window_size}'] = new_df.iloc[:,0].rolling(window_size, center=True).mean()
new_df[f'window_{window_size*10}'] = new_df.iloc[:,0].rolling(window_size*10, center=True).mean()
new_df[f'window_{window_size*50}'] = new_df.iloc[:,0].rolling(window_size*50, center=True).mean()
new_df[f'window_{window_size*100}'] = new_df.iloc[:,0].rolling(window_size*100, center=True).mean()
plot_2 = px.line(new_df)
plot_2.show()
```

```python
new_df_2 = df.resample('w').mean()

window_size = 10
new_df_2[f'window_{window_size}'] = new_df_2.iloc[:,0].rolling(window_size, center=True).mean()
new_df_2[f'window_{window_size*10}'] = new_df_2.iloc[:,0].rolling(window_size*10, center=True).mean()
new_df_2[f'window_{window_size*50}'] = new_df_2.iloc[:,0].rolling(window_size*50, center=True).mean()
new_df_2[f'window_{window_size*100}'] = new_df_2.iloc[:,0].rolling(window_size*100, center=True).mean()
plot_3 = px.line(new_df_2)
plot_3.show()
```
