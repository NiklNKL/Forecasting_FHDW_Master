```python
import pandas as pd
import plotly_express as px
df = pd.read_csv("C:\Users\janni\Desktop\FHDW\Forecasting\Python\data\Ch03_Earthquake_database.csv")
new_df = df.copy()
new_df['datetime'] = pd.to_datetime(new_df['Date'] + ' ' + new_df['Time'])
new_df = new_df.set_index("datetime").resample('A').mean()
filtered_df = new_df["Magnitude"] >=7
line = px.line(filtered_df, y=filtered_df["Magnitude"])
line.show()
```
