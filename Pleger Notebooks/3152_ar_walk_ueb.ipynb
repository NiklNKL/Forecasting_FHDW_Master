```python
%reset -f

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
fn1 = 'data/ar_walk01.csv'
fn2 = 'data/ar_walk02.csv'
fn3 = 'data/ar_walk03.csv'
```

```python
ds = pd.read_csv(fn1)
display(ds)
```

```python
from statsmodels.tsa.arima.model import ARIMA
import numpy as np
import pandas as pd

import statsmodels.api as sm

# Define the range of orders to try
p_values = range(0, 7)
d_values = range(0, 7)
q_values = range(0, 7)

# Initialize variables to store the best order and AIC value
best_order = None
best_aic = np.inf

# Iterate over all possible orders
for p in p_values:
    for d in d_values:
        for q in q_values:
            try:
                # Fit the ARIMA model
                model = ARIMA(ds, order=(p, d, q))
                results = model.fit()

                # Calculate the AIC value
                aic = results.aic

                # Update the best order and AIC value if necessary
                if aic < best_aic:
                    best_order = (p, d, q)
                    best_aic = aic
            except:
                continue

# Print the best order and AIC value
print("Best Order:", best_order)
print("Smallest AIC Value:", best_aic)
```

```python
# Remove the unused import
from numba import jit
from joblib import Parallel, delayed
from sklearn.model_selection import GridSearchCV
from statsmodels.tsa.arima.model import ARIMA
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV
from statsmodels.tsa.arima.model import ARIMA
from numba import jit
import numpy as np
import pandas as pd

import statsmodels.api as sm

@jit
def find_best_order(ds, p_values, d_values, q_values):
    # Define the parameter grid for grid search
    param_grid = {'p': p_values, 'd': d_values, 'q': q_values}

    # Define the ARIMA model
    model = ARIMA(ds, order=(0, 0, 0))

    # Perform grid search with parallel computing
    grid_search = GridSearchCV(estimator=model, param_grid=param_grid, scoring='aic', cv=3, n_jobs=-1)
    grid_search.fit(ds)

    # Get the best order and AIC value
    best_order = grid_search.best_params_
    best_aic = grid_search.best_score_

    # Return the best order and AIC value
    return best_order, best_aic

# Define the range of orders to try
p_values = range(0, 7)
d_values = range(0, 7)
q_values = range(0, 7)

# Call the function with the necessary arguments
best_order, best_aic = find_best_order(ds, p_values, d_values, q_values)

# Print the best order and AIC value
print("Best Order:", best_order)
print("Smallest AIC Value:", best_aic)
```
