## PDF-Ausgabe

PDF-Dateien erfordern Latex. Latex sollte auf jedem Rechner installiert sein, der mit mathematischen Formeln in Berhrung kommt. Unter Windows funktioniert zum Beispiel TexStudio (<https://www.texstudio.org/>).



Bei Linux bewhrt sich *MikTex*.

Mgliche Probleme bei der Direktausgabe:

* Eingebettete Bilder werden nicht erkannt und Ausgabe bricht mit Fehlermeldung ab (Issue von Januar 2022, Problem scheint noch nicht gelst).
* Markdown-Export funktioniert, aber die Links zu den eingebetteten Bildern werden nicht korrigiert.

Notlsung:

* Statt ber file/export gelingt die Ausgabe print/pdf fast immer, wenn auch nicht ganz so ansprechend wie der Direktexport zu pdf. 

Beste Lsung:

*Pandoc*, das freie schweizer Taschenmesser fr die Konversion von Textformaten: <https://pandoc.org/>

Aufruf: `pandoc meineDatei.ipynb -o meineDatei.pdf`

Sofern der Ausdruck Codezeilen nicht deutlich vom Text abhebt, sollten Begrenzungslinien (\<hr\>) eingefgt werden, ggf. auch zustliche Leerzeilen ("\&nbsp;", siehe Quelltext).
