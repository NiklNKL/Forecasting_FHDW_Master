# Simple plots

Mean annual temperature of the past years in Wien (AT) and Germany.

Data sources.

* https://de.wikipedia.org/wiki/Zeitreihe_der_Lufttemperatur_in_Deutschland
* https://www.wien.gv.at/statistik/lebensraum/tabellen/temperatur-zr.html

The data is given below. Later we will extract the data directly from the HTML pages.

```python
t_de = [ 9.6, 9.1, 8.7, 10.3, 9.9, 9.6, 9.6, 10.45, 10.3, 10.4 ]
t_at = [11.1, 11.3, 10.9, 12., 12.1, 11.5, 11.6, 12.4, 12.4, 11.9]
years = list(range(2011,2021))
```

```python
import matplotlib.pyplot as plt

plt.plot(t_de)
```

```python
plt.plot(t_de)
plt.plot(t_at)
plt.show()
```

```python
plt.plot(years, t_de, label = 'DE')
plt.plot(years,t_at,label = 'AT')
plt.title('Mean Temperature in AT and DE')
plt.xlabel('Year')
plt.ylabel('Mean Temperature')
# plt.legend()
plt.show()
```

```python
#plt.figure(figsize=(8,2), dpi=200)
plt.plot(years, t_de, label = 'DE')
plt.plot(years,t_at,label = 'AT')
plt.title('Mean Temperature in AT and DE')
plt.xlabel('Year')
plt.ylabel('Mean Temperature')
plt.legend()
plt.ylim(5,15)
plt.grid()
plt.show()
```

```python
plt.fill_between(years, t_de, alpha = 0.3)
plt.fill_between(years,t_at, alpha = 0.3)
plt.ylim(0,)
plt.show()
```

```python
plt.fill_between(years, t_de, 8, alpha = 0.3)
plt.fill_between(years,t_at, 8, alpha = 0.3)
plt.ylim(0,)
plt.show()
```

```python
plt.fill_between(years, t_de, t_at, alpha = 0.3)
plt.ylim(0,)
plt.show()
```

```python
plt.scatter(years, t_de, s=10)
#plt.plot(years, t_de)
plt.show()
```

```python
#plt.scatter(years, t_de)
plt.plot(years, t_de, 'ro--')
plt.show()
```

```python
plt.plot(years, t_de, color='red', marker = 'o', markersize=10, linewidth=3, linestyle='--')
plt.show()
```

```python
plt.stem(years, t_de)
plt.show()
```

```python
plt.bar(years, t_de)
plt.show()
```

```python
plt.title('Frequency of mean temperature intervals')
plt.hist(t_de, bins=3)
plt.show()
```

## Object-Oriented Interface




Hierarchy of `Figure`, `Artist`, `Axes` and `Axis`

```python

fig, ax1 = plt.subplots()

ax1.set_ylabel("Temperature Germany")
ax1.set_xlabel("year")
ax1.plot(years, t_de, "blue")

ax2 = ax1.twinx() # create another y-axis sharing a common x-axis

ax2.set_ylabel("Temperature in Vienna")
ax2.set_xlabel("year")
ax2.plot(years, t_at, "green")

fig.set_size_inches(7,5)
fig.set_dpi(100)
fig.legend(['Temperature DE', 'Temperature AT'],
          loc="upper left") #, bbox_to_anchor=(0.13,0.85))


plt.show()
```
