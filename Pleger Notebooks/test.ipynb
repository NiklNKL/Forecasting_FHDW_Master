```python
import numba as nb
import numpy as np
import time
@nb.jit(nopython=True)
def sum_array_numba(arr):
    total = 0
    for i in range(len(arr)):
        total += arr[i]
    return total

def sum_array(arr):
    total = 0
    for i in range(len(arr)):
        total += arr[i]
    return total

# Create a large array
arr = np.random.randint(0, 100, size=10**6)

# Calculate the sum using Numba
start_time_numba = time.time()
result_numba = sum_array_numba(arr)
end_time_numba = time.time()

# Calculate the sum without Numba
start_time = time.time()
result = sum_array(arr)
end_time = time.time()

# Print the results and execution times
print("Result with Numba:", result_numba)
print("Execution time with Numba:", end_time_numba - start_time_numba, "seconds")
print("Result without Numba:", result)
print("Execution time without Numba:", end_time - start_time, "seconds")
```
